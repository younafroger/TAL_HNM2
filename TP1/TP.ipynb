```python
with open ("pg13846.txt", "r") as infile : # Discours sur la methode
  content_1 = infile.read ()

with open ("pg4300.txt", "r") as infile : # Ulysses
  content_2 = infile.read ()




def read_file ( path_to_file ) :
  with open ( path_to_file , "r", encoding ="utf-8") as infile :
    content = infile . read ()
  return content

content_discours = read_file ("pg13846.txt")# Discours de la Methode
content_ulysses = read_file ("pg4300.txt")# Ulysses

print(content_discours[:100])
print(content_ulysses[:100])
```

```python
def split_into_words ( content_unsplit ) :
  #we split
  words_list = content_unsplit.split()
  return words_list
```

```python
def get_lengths_of_words (word_list):
  lengths_dic = dict()
  for word in word_list :
    length = len(word)
    if length not in lengths_dic :
      lengths_dic[length] = 1
    else :
      lengths_dic[length] += 1
  #print(lengths_dic)
  return lengths_dic
```

```python
def print_stats (lengths_dic):
  num_occurrences = dict()
  for i in range (1,30) :
    if i in lengths_dic :
      num_occurrences[i] = lengths_dic[i]
    else :
      num_occurrences[i] = 0
  sum_words = 0
  for i in range (1,30) :
    sum_words = sum_words + num_occurrences[i]
  for i in range (1,30) :
    ratio = round((num_occurrences[i]*100)/sum_words,2)
    print ( f" Size : {i} \ Occurrences : {num_occurrences[i]} \ Ratio : {ratio}%")
```

```python
liste_discours = split_into_words(content_discours)
liste_ulysses = split_into_words(content_ulysses)

dic_discours = get_lengths_of_words(liste_discours)
dic_ulysses = get_lengths_of_words(liste_ulysses)

print_stats(dic_discours)
print_stats(dic_ulysses)
```

```python
import matplotlib . pyplot as pyplot # import avec alias

def plot_books ( books ) :
  for book in books:
    lengths_dic = get_lengths_of_words(book)
    lengths_ordered_list = []
    for i in range (30) :
      if i in lengths_dic :
        lengths_ordered_list.append(lengths_dic[i])
      else :
        lengths_ordered_list.append (0)
    pyplot.plot (lengths_ordered_list) #we " draw "
  pyplot.show () #we " display "

plot_books ([ liste_discours , liste_ulysses ])
```

```python
def plot_books_change ( books ) :
  for book in [liste_discours, liste_ulysses]:
    taille= len(book)
    lengths_dic = get_lengths_of_words(book)
    lengths_ordered_list = []
    for i in range (30) :
      if i in lengths_dic :
        lengths_ordered_list.append(((lengths_dic[i]*100)/taille)/100)
      else :
        lengths_ordered_list.append (0)
    pyplot.plot (lengths_ordered_list) #we " draw "
  pyplot.show () #we " display "

plot_books_change([ liste_discours , liste_ulysses ])
```
